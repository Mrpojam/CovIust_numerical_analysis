# Stat analysis

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.regplot(x=data['total_cases_per_million'], y=data['total_deaths_per_million'])
plt.savefig('../Plots/case_death.png')
```

```python
sns.scatterplot(x=data['total_cases_per_million'], y=data['total_deaths_per_million'], size=data['gdp_per_capita'])
plt.savefig('../Plots/case_death_gdp.png')
```

```python
sns.set(rc={'figure.figsize':(15.7,15.27)})
```

```python
sns.scatterplot(x=data['population_density'], y=data['hospital_beds_per_thousand'], hue=data['location'], size=data['extreme_poverty'])
plt.savefig('../Plots/case_hospital_povery.png')
```
