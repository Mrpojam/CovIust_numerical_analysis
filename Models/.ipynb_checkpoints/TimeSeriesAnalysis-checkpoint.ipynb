# Panel Data analysis
Panel data is also a time based dataset.

The difference is that, in addition to time series, it also contains one or more related variables that are measured for the same time periods.

Typically, the columns present in panel data contain explanatory variables that can be helpful in predicting the Y, provided those columns will be available at the future forecasting period.

```python
def plot_df(df, x, y, title="", file_name = "", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.savefig("../Plots/" + file_name)
    plt.show()
```

## Iran

```python
plot_df(ir_data, x=ir_data.index, y=ir_data['new_deaths'], title='Dayly covid-19 deaths timeseries for last 2 years.', file_name='ir_death_ts.png')
plot_df(ir_data, x=ir_data.index, y=ir_data['new_cases'], title='Dayly covid-19 cases timeseries for last 2 years.', file_name='ir_cases_ts.png')    
```
