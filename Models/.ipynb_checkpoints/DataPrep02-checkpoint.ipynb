# Data Cleaning

```python
 a = data.isnull().sum()
print(a)
```

```python
data.columns
```

```python
usefull_columns_stats = ['date', 'continent', 'location', 'total_cases', 'total_deaths', 'total_cases_per_million', 'total_deaths_per_million', 'total_vaccinations_per_hundred', 'population', 'population_density', 'diabetes_prevalence', 'female_smokers', 'male_smokers', 'gdp_per_capita', 'extreme_poverty', 'hospital_beds_per_thousand', 'life_expectancy', 'human_development_index']
```

```python
data = data[usefull_columns_stats]
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.sort_values(by='date', ascending=False, inplace=True)
data
```

```python
data = data[data['date'] == '2022-04-18']
```
